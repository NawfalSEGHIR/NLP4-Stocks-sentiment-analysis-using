### Data Cleaning and Preprocessing

```python
df.isna().any() # findingb all the NaN values in our Data set
```

```python
# Dropping NaN values and check the updated number of rows (some days we havent colected all the data required, so we will delete the entire day from the data set)
df.dropna(inplace=True)
print(df.shape)
```

```python
df_copy = df.copy() # we create a copy  of our original data set in order to make modification safely
```

```python
df_copy.reset_index(inplace=True)
```

```python
df_copy.head()
```

```python
import mitosheet
mitosheet.sheet(df_copy, view_df=True)
```

```python
# Splitting the dataset into train/test set

train = df_copy[df_copy['Date'] < '20150101']
test = df_copy[df_copy['Date'] > '20141231']
print('Train size: {}, Test size: {}'.format(train.shape, test.shape))
```

```python
# Splitting the dataset

y_train = train['Label']
train = train.iloc[:, 3:28] # from collumn 3 to 28 we have the list of our stock headlines per day
y_test = test['Label']
test = test.iloc[:, 3:28]
```

```python
# Importing essential libraries for performing Natural Language Processing on our dataset

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords # stopwors are words without much meaning added to a sentence (example: the, he, have... etc)
from nltk.stem import PorterStemmer # Porter Stemmer is the inventor of the Stemming algorithm
```

```python
# Removing punctuation and special character from the text

train.replace(to_replace='[^a-zA-Z]', value=' ', regex=True, inplace=True)
test.replace(to_replace='[^a-zA-Z]', value=' ', regex=True, inplace=True)
```

```python
# Renaming columns

new_columns = [str(i) for i in range(0,25)]
train.columns = new_columns
test.columns = new_columns
```

```python
train.head() #notice that columns are now named from 0 to 24 and no more punctuation on our Headlines
```

```python
import mitosheet
mitosheet.sheet(train.head() #notice that columns are now named from 0 to 24 and no more punctuation on our Headlines, view_df=True)
```

```python
# Converting the entire text to lower case

for i in new_columns:
  train[i] = train[i].str.lower()
  test[i] = test[i].str.lower()
```

```python
train.head() # now all the text is in lower case
```

```python
import mitosheet
mitosheet.sheet(train.head() # now all the text is in lower case, view_df=True)
```

```python
# Joining all the columns to create a text like day by day (row by row)

train_headlines = []
test_headlines = []

for row in range(0, train.shape[0]):
  train_headlines.append(' '.join(str(x) for x in train.iloc[row, 0:25]))

for row in range(0, test.shape[0]):
  test_headlines.append(' '.join(str(x) for x in test.iloc[row, 0:25]))
```

```python
train_headlines[1]
```

```python
test_headlines[1]
```

```python
# Creating corpus of train dataset

ps = PorterStemmer() # here we will use the stemming algo instead of the lemmatization
train_corpus = []

for i in range(0, len(train_headlines)):
  
  # Tokenizing the Headlines by words = splitting the text into words
  words = train_headlines[i].split()

  # Removing the stopwords
  words = [word for word in words if word not in set(stopwords.words('english'))]

  # Stemming the words
  words = [ps.stem(word) for word in words]

  # Joining the stemmed words
  headline = ' '.join(words)

  # Building a corpus of Headlines
  train_corpus.append(headline)
```

```python
train_corpus[0:5] # view of the 5 first arrows
```

```python
# same thing here, we create a corpus of test dataset
test_corpus = []

for i in range(0, len(test_headlines)):
  
  # Tokenizing the news-title by words
  words = test_headlines[i].split()

  # Removing the stopwords
  words = [word for word in words if word not in set(stopwords.words('english'))]

  # Stemming the words
  words = [ps.stem(word) for word in words]

  # Joining the stemmed words
  headline = ' '.join(words)

  # Building a corpus of news-title
  test_corpus.append(headline)
```

```python
test_corpus[0:5]
```

```python
down_words = [] #down words are words making stocks going down
for i in list(y_train[y_train==0].index):
  down_words.append(train_corpus[i])

up_words = []
for i in list(y_train[y_train==1].index): #up words are words making stocks going up
  up_words.append(train_corpus[i])
```

```python
# Creating the Bag of Words model - features extraction for text modeling

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, ngram_range=(2,2))

X_train = cv.fit_transform(train_corpus).toarray()
X_test = cv.transform(test_corpus).toarray()
```

## Model Building 
(We will compare the performance of 3 different models: Linear Regression, Random Forest & Naive Bayes, and the best one will be chosen for our Predictions)

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)
```

```python
lr_y_pred = lr_classifier.predict(X_test)
```

```python
# measure the Accuracy, Precision and Recall of this model

from sklearn.metrics import accuracy_score, precision_score, recall_score

score1 = accuracy_score(y_test, lr_y_pred)
score2 = precision_score(y_test, lr_y_pred)
score3 = recall_score(y_test, lr_y_pred)

print("---- Scores ----")
print("Accuracy score is: {}%".format(round(score1*100,2))) # here we choose to ahve an output in % format
print("Precision score is: {}".format(round(score2,2)))
print("Recall score is: {}".format(round(score3,2)))
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
lr_cm = confusion_matrix(y_test, lr_y_pred)

lr_cm
```

```python
# Plotting the confusion matrix

plt.figure(figsize=(10,7))
sns.heatmap(data=lr_cm, annot=True, cmap="Blues", xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])

plt.xlabel('Predicted values')
plt.ylabel('Actual values')

plt.title('Confusion Matrix for Logistic Regression Algorithm')

plt.show()
```

### 2. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy') # with 100 trees in our Forest, and Entropy as Impurity metric
rf_classifier.fit(X_train, y_train)
```

```python
rf_y_pred = rf_classifier.predict(X_test)
```

```python
# Accuracy, Precision and Recall measurement 

score1 = accuracy_score(y_test, rf_y_pred)
score2 = precision_score(y_test, rf_y_pred)
score3 = recall_score(y_test, rf_y_pred)

print("---- Scores ----")
print("Accuracy score is: {}%".format(round(score1*100,2)))
print("Precision score is: {}".format(round(score2,2)))
print("Recall score is: {}".format(round(score3,2)))
```

##### Note: Logistic Regression has slightly better performance in this case, with a Precision score of 0.87

```python
# Confusion Matrix

rf_cm = confusion_matrix(y_test, rf_y_pred)

rf_cm
```

```python
# Plotting the confusion matrix

plt.figure(figsize=(10,7))
sns.heatmap(data=rf_cm, annot=True, cmap="binary", xticklabels=['Down', 'Up'], yticklabels=['Down', 'Up'])

plt.xlabel('Predicted values')
plt.ylabel('Actual values')

plt.title('Confusion Matrix for Random Forest Algorithm')

plt.show()
```
